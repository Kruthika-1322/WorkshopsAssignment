```python
def crop_yield(region,weights):
    '''
    This function computes crop yield of a given region
    Parameters: (region as list, weights as list)
    Return: float value
    '''
    result=0
    for x,w in zip(region,weights):
        result=result+x*w
    return result
print("The expected yeild of apples in the region is: ",crop_yield(hoenn,weights))
```

```python
import urllib.request
url='https://raw.githubusercontent.com/chidanandamurthy/dataAnalysis/main/climate.txt'
urllib.request.urlretrieve(url,'climate.txt')
```

```python
climate_data=np.genfromtxt('climate.txt',delimiter=',',skip_header=1)
```

```python
climate_data
```

```python
climate_data.shape
```

```python
yeilds=climate_data@weights
```

```python
yeilds.shape
```

```python
climate_results=np.concatenate((climate_data,yeilds.reshape(10000,1)),axis=1)
climate_results
```

```python
np.savetxt('climate_results.txt',climate_results,fmt='%2.f',delimiter=',',
           header='Temperature,Rainfall,Humidity,Yeild_apples',comments='')
```
