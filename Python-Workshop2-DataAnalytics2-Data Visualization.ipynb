```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
years = [y for y in range(2000,2012)]
print(years)
apples= [0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936,0.937,0.9375,0.9372,0.939]
oranges=[0.962,0.941,0.930,0.923,0.918,0.908,0.907,0.904,0.901,0.898,0.9,0.869]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)

plt.xlabel('Year')
plt.ylabel('yield(tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(years,apples, marker='o')
plt.plot(years,oranges,marker='x')

plt.xlabel('Year')
plt.ylabel('yield(tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(years,apples,marker='s',c='b',ls='-',lw=2,ms=5,mew=5,mec='pink',alpha=.3)
plt.plot(years,oranges,marker='x',c='r',ls='--',lw=3,ms=10,alpha=.3)

plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')

plt.title('Crop yield in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(years,apples,'s-b')
plt.plot(years,oranges,'o--r')

plt.xlabel('Year')
plt.ylabel('yield(tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples','Oranges']);
```

```python
flower_df=sns.load_dataset("iris")
flower_df
```

```python
flower_df.species.unique()
```

```python
plt.plot(flower_df.sepal_length,flower_df.sepal_width);
```

```python
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width)
```

```python
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width, hue=flower_df.species,s=100)
```

```python
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width, hue=flower_df.species,s=100)
```

```python
plt.title("Distribution of sepal width")
plt.hist(flower_df.sepal_width)
```

```python
plt.hist(flower_df.sepal_width,bins=8)
```

```python
import numpy as np
plt.hist(flower_df.sepal_width,bins=np.arange(2,5,0.25));
```

```python
plt.hist(flower_df.sepal_width,bins=[1,3,4,4.5])
```

```python
setosa_df=flower_df[flower_df.species=='setosa']
versicolor_df=flower_df[flower_df.species=='versicolor']
virginica_df=flower_df[flower_df.species=='verginica']
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
```

```python
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],bins=np.arange(2,5,.25),stacked=True,alpha=0.4);
plt.legend(['Setosa','Versicolor','Virginica']);
```
